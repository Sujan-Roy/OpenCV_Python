# What is Morphological transformation

Morphological transformation is a set of image processing techniques used to manipulate the shape, size, and structure of objects in an image. These techniques are based on the mathematical theory of morphology, which deals with the structure and form of objects.

# Classification of Morphological transformation 

Morphological transformation techniques can be classified into two categories: 
1. binary and 
2. grayscale.

# Binary Morphological transformation
Binary morphological transformations are used to process binary images, where each pixel is either black or white.The most common binary morphological transformations are erosion, dilation, opening, and closing.  

# Grayscale Morphological transformation
Grayscale morphological transformations are used to process grayscale images, where each pixel has a value between 0 and 255 representing its brightness. The most common grayscale morphological transformations are top-hat, bottom-hat, and gradient. Top-hat enhances the small features in an image, while bottom-hat enhances the larger features. Gradient enhances the edges of objects in an image.

# Application of Morphological transformation
Morphological transformations can be applied in a variety of image processing tasks, including image segmentation, noise reduction, edge detection, and feature extraction. They are particularly useful in tasks where the shape and structure of objects in an image are important, such as medical imaging and industrial inspection.

# Sample code for Morphological transformation
Here I will expalin the basic code of Morphological transformation where I used only one image as a input

First, import the following library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
```

Take a image as a Iuput. Here I used cat image as a Input.

To show the output of Morphological transformation, I defined a user define function ( named as catImageShow)

```python
def catImageShow(imagetitle, image):
    imagevaraiable= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(4,3))
    plt.imshow(imagevaraiable)
    plt.title(imagetitle)
    plt.show()
```

## Erosion morphological transformation operation
Erosion is a morphological transformation operation in image processing that is used to remove small or thin structures from an image. It works by convolving a structuring element, which is a small binary image, with the input image. The structuring element is moved over the input image, and the output image is generated by taking the minimum pixel value of all the pixels in the input image that are overlapped by the structuring element.

# Application of Erosion morphological transformation operation
The erosion operation is often used to remove noise from binary images or to separate connected regions in an image. It can also be used in combination with other morphological operations, such as dilation and opening, to perform more complex image processing tasks.

# Syntax of Erosion morphological transformation operation:

erode (src, dst, kernel, anchor, iterations, borderType, borderValue)

Parameters:

1. src: Input image
2. dst: Output image
3. kernel: The structuring element used for erosion.
4. anchor: Anchor position inside the structuring element. The default value is [-1, -1} signifying position as the center of the structuring element.

5. iterations: Number of times erosion is applied.
6. borderType: Type of border (BORDER_CONSTANT, BORDER_REPLICATE, etc.)
7. borderValue: Border value

In this code, I used only 3 parameters. That are src, none, iterations.

```python
image = cv2.imread(imageArgs["image"])
imagegray= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
catImageShow("Cat image", imagegray)
for i in range(0,5):
    imageErosion = cv2.erode(imagegray,None,iterations =i+1) 
    catImageShow("Image imageErosion number is {}".format(i+1),imageErosion)
```

# Dilation morphological transformation operation
Dilation is the opposite of Erosion, instead of shrinking it expands the foreground object.
It involves modifying the shape of an object in an image by expanding its boundaries or making it thicker.
The effect of dilation on an image depends on the size and shape of the structuring element used. If the structuring element is small, the dilation will produce small changes in the image. If the structuring element is large, the dilation will produce larger changes in the image.

# Application of Dilation morphological transformation operation
Dilation is often used to fill small gaps or holes in objects in an image, to smooth the boundaries of objects, or to increase the size of objects in an image. It can also be used as a preprocessing step for other image processing operations such as edge detection, segmentation, and feature extraction.

# Syntax of Dilation morphological transformation operation:
dilate (src, dst, kernel, anchor, iterations, borderType, borderValue)


```python
image = cv2.imread(imageArgs["image"])
imagegray= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
catImageShow("Cat image", imagegray)
for i in range(0,5):
    imageDilate = cv2.dilate(imagegray,None,iterations =i+1)
    catImageShow("Image Dilate number is {}".format(i+1),imageDilate)
```

# Opening morphological transformation operation
This operation is a combination of erosion followed by dilation. It is usually used for removing internal noise present inside an image. 
# Syntax
cv2.morphologyEx(src, cv2.MORPH_OPEN, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]])

```python
image = cv2.imread(imageArgs["image"])
imagegray= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
kernelSize = [(3,3),
             (7,7),
             (9,9)]
for i in kernelSize:
    k=cv2.getStructuringElement(cv2.MORPH_RECT, i)
    mo = cv2.morphologyEx(imagegray,cv2.MORPH_OPEN,k)
    catImageShow("Opening Output : ({}, {})".format(kernelSize[0],kernelSize[1]),mo)
```

# Closing morphological transformation operation
This operation is a combination of dilation followed by erosion. It can be used to close small holes inside the foreground objects.

# Syntax
cv2.morphologyEx(src, cv2.MORPH_CLOSE, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]])

```python
image = cv2.imread(imageArgs["image"])
imagegray= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
kernelSize = [(3,3),
             (7,7),
             (9,9)]
for i in kernelSize:
    k=cv2.getStructuringElement(cv2.MORPH_RECT, i)
    mo = cv2.morphologyEx(imagegray,cv2.MORPH_CLOSE,k)
    catImageShow("Closing Output : ({}, {})".format(kernelSize[0],kernelSize[1]),mo)
```

# Gradient
This operation is the difference between dilation and erosion of an image. It can be used to extract edges from an image.
# Syntax
cv2.morphologyEx(src, cv2.MORPH_GRADIENT, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]])

```python
image = cv2.imread(imageArgs["image"])
imagegray= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
kernelSize = [(3,3),
             (5,5),
             (7,7)]

#catimageShow("Cat image", imagegray)
for i in kernelSize:
    k=cv2.getStructuringElement(cv2.MORPH_RECT, i)
    mo = cv2.morphologyEx(imagegray,cv2.MORPH_GRADIENT,k)
    catImageShow("Gradient output : ({}, {})".format(kernelSize[0],kernelSize[1]),mo)
```

# Summary
In all these functions, the src argument is the input image, and the kernel argument is the structuring element used for the operation. The other arguments are optional and can be used to specify the output image, the anchor point of the kernel, the number of iterations, the border type, and the border value.
