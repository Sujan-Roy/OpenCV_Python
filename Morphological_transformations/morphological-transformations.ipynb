```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import warnings

%matplotlib inline
```

```python
imageArgs = {"image": "../input/catmor/inputcat1.jpg"}
```

```python
def catimageShow(imagetitle, image):
    imagevaraiable= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(4,3))
    plt.imshow(imagevaraiable)
    plt.title(imagetitle)
    plt.show()
```

```python
image = cv2.imread(imageArgs["image"])
imagegray= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
catimageShow("Cat image", imagegray)
for i in range(0,5):
    imageEroded = cv2.erode(imagegray,None,iterations =i+1)
    catimageShow("Image Eroded number is {}".format(i+1),imageEroded)
```

```python
image = cv2.imread(imageArgs["image"])
imagegray= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
catimageShow("Cat image", imagegray)
for i in range(0,5):
    imageDilate = cv2.dilate(imagegray,None,iterations =i+1)
    catimageShow("Image Dilate number is {}".format(i+1),imageDilate)
```

```python
image = cv2.imread(imageArgs["image"])
imagegray= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
kernelSize = [(3,3),
             (5,5),
             (7,7)]

#catimageShow("Cat image", imagegray)
for i in kernelSize:
    k=cv2.getStructuringElement(cv2.MORPH_RECT, i)
    mo = cv2.morphologyEx(imagegray,cv2.MORPH_OPEN,k)
    catimageShow("Opening info : ({}, {})".format(kernelSize[0],kernelSize[1]),mo)
   # imageOpen = cv2.morphologyEx(imagegray,None,iterations =i+1)
    #catimageShow("Image Dilate number is {}".format(i+1),imageOpen)
```
